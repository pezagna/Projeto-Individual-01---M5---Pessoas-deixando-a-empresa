## **Realizando a importao e limpeza dos dados.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_excel("EXERCISE 8.5.xlsx")
df
```

```python
# Eliminar colunas com valores NaN
df = df.dropna(axis=1, how='all')

# Eliminar linhas com valores NaN
df = df.dropna(axis=0, how='all')
# Redefinir o ndice
df = df.reset_index(drop=True)
# Exibir o DataFrame resultante
print(df)
```

```python
# Excluir as linhas 1 e 2
df = df.drop([0, 1])

# Renomear as colunas
df.columns = ['Dados', 'Ability to impact', 'HRBP', 'Exit survey']

# Redefinir o ndice
df = df.reset_index(drop=True)

# Exibir o resultado
print(df)
```

Observaes a partir dos dados:

A categoria "Career advancement" tem a maior pontuao em "Ability to impact" e "HRBP", o que indica que a progresso de carreira  vista como um fator significativo na satisfao dos funcionrios e  uma rea em que a empresa est desempenhando bem.
A categoria "Workload" tem uma pontuao relativamente baixa em "Ability to impact" e  mencionada no "Exit survey", sugerindo que a carga de trabalho pode ser um desafio e pode afetar negativamente a satisfao dos funcionrios.

A categoria "Pay" tem uma pontuao moderada em "Ability to impact" e tambm  mencionada no "Exit survey", indicando que a remunerao pode ser um fator relevante na satisfao dos funcionrios, mas no  o nico determinante.

## **Grfico de Barras:** 

**Vantagens:**
Permite comparar facilmente as pontuaes de "Ability to impact", "HRBP" e "Exit survey" para cada categoria, facilitando a identificao de diferenas exibindo a distribuio das pontuaes em cada categoria.

**Deficincias:**
Pode ser difcil visualizar as diferenas sutis entre as barras quando h muitas categorias e valores baixos.

```python
# Definir a largura das barras
bar_width = 0.25

# Criar uma sequncia de ndices para posicionar as barras
index = np.arange(len(df))
plt.figure(figsize=(10, 9))
# Plotar as barras de Exit survey
plt.barh(index, df['Exit survey'], color='lightcoral', height=bar_width, label='Exit survey')

# Plotar as barras de HRBP
plt.barh(index + bar_width, df['HRBP'], color='lightgreen', height=bar_width, label='HRBP')

# Plotar as barras de Ability to impact
plt.barh(index + (2 * bar_width), df['Ability to impact'], color='skyblue', height=bar_width, label='Ability to impact')

# Personalizar o eixo y
plt.yticks(index + bar_width, df['Dados'])

# Personalizar o eixo x
plt.xlabel('Pontuao')
plt.title('Satisfao dos funcionrios por categoria')

# Adicionar uma legenda
plt.legend()

# Adicionar os nmeros das barras
for i, value in enumerate(df['Exit survey']):
    plt.text(value, i, str(value), ha='left', va='center', color='black')

for i, value in enumerate(df['HRBP']):
    plt.text(value, i + bar_width, str(value), ha='left', va='center', color='black')

for i, value in enumerate(df['Ability to impact']):
    plt.text(value, i + (2 * bar_width), str(value), ha='left', va='center', color='black')

# Mostrar o grfico
plt.show()
```

### **Grfico de Disperso.** 

**Vantagens:**
Mostra a relao entre "Ability to impact" com "HRBP" e "Exit survey" para cada categoria.
Permite identificar padres ou clusters nos dados.

**Deficincias:**
Pode ser difcil interpretar as pontuaes exatas em cada ponto de disperso.
No fornece uma viso clara da distribuio das pontuaes.


```python
# Grfico de Disperso 1
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Exit survey', y='Ability to impact', hue='Dados', s=100, palette='viridis')
plt.xlabel('Exit Survey')
plt.ylabel('Ability to impact')
plt.title('Relao entre Ability to impact e Exit Survey')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Grfico de Disperso 2

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='HRBP', y='Ability to impact', hue='Dados', s=100, palette='viridis')
plt.xlabel('HRBP')
plt.ylabel('Ability to impact')
plt.title('Relao entre Ability to impact e HRBP')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### **Grfico de Linhas:**
**Vantagens:**
Permite acompanhar as tendncias das pontuaes ao longo das categorias.
Destaca facilmente as categorias com pontuaes mais altas ou mais baixas.

**Deficincias:**
No fornece uma representao visual precisa das pontuaes individuais em cada categoria.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Dados', y='Ability to impact', marker='o', label='Ability to impact')
sns.lineplot(data=df, x='Dados', y='HRBP', marker='o', label='HRBP')
sns.lineplot(data=df, x='Dados', y='Exit survey', marker='o', label='Exit survey')
plt.xticks(rotation=35)
plt.xlabel('Dados')
plt.ylabel('Pontuao')
plt.title('Satisfao dos funcionrios por categoria')
plt.legend()
plt.show()
```

A visualizao que melhor representa os dados depende do objetivo da anlise. Se o objetivo for comparar as mtricas para cada categoria, o grfico de barras agrupadas pode ser mais adequado. Se o objetivo for destacar a composio das pontuaes em cada categoria, o grfico de barras empilhadas pode ser prefervel. J se o objetivo for identificar possveis relaes entre as mtricas de satisfao dos funcionrios, o grfico de disperso pode ser mais adequado.
