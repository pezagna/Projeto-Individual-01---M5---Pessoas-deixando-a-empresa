Neste gráfico de bolhas, estamos apresentando a satisfação dos funcionários por categoria, representada pelos eixos "Ability to impact" (Capacidade de impacto) e "Frequency" (Frequência). Cada bolha no gráfico representa uma categoria específica, como treinamento, conflito com outros, falta de reconhecimento, carga de trabalho, progressão de carreira, remuneração, tipo de trabalho, entre outros. O tamanho das bolhas representa as pontuações em "HRBP" (Pontuação do HRBP) e "Exit survey" (Pesquisa de saída), proporcionando uma visão do tamanho relativo da satisfação geral dos funcionários em cada categoria.

Obs: Esses dados são relativamente fáceis de falar e interpretar em sua forma atual pois a exibição dos dados e a visualização estão bem destacadas.

**2. Observações:**
O gráfico de bolhas visualmente destaca as diferentes categorias. Seus tamanhos correspondem à satisfação geral dos funcionários. Isso facilita a identificação das categorias com maior ou menor satisfação.

**Vantagens do gráfico de bolhas:**

Destaca visualmente as diferentes categorias e sua satisfação relativa.

**Desvantagens do gráfico de bolhas:**

A comparação precisa das áreas das bolhas requer atenção extra.
Pode ser difícil para um usuário iniciante conseguir interpretar rapidamente um gráfico de bolhas.
Dependente do tamanho das bolhas o gráfico pode ser tornar confuso.

## **Realizando a importação e limpeza dos dados.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df = pd.read_excel("EXERCISE 8.5.xlsx")
df
```

```python
# Eliminar colunas com valores NaN
df = df.dropna(axis=1, how='all')

# Eliminar linhas com valores NaN
df = df.dropna(axis=0, how='all')
# Redefinir o índice
df = df.reset_index(drop=True)
# Exibir o DataFrame resultante
print(df)
```

```python
# Excluir as linhas 1 e 2
df = df.drop([0, 1])

# Renomear as colunas
df.columns = ['Dados', 'Ability to impact', 'HRBP', 'Exit survey']

# Redefinir o índice
df = df.reset_index(drop=True)

# Exibir o resultado
print(df)
```

Observações a partir dos dados:

A categoria "Career advancement" tem a maior pontuação em "Ability to impact" e "HRBP", o que indica que a progressão de carreira é vista como um fator significativo na satisfação dos funcionários e é uma área em que a empresa está desempenhando bem.
A categoria "Workload" tem uma pontuação relativamente baixa em "Ability to impact" e é mencionada no "Exit survey", sugerindo que a carga de trabalho pode ser um desafio e pode afetar negativamente a satisfação dos funcionários.

A categoria "Pay" tem uma pontuação moderada em "Ability to impact" e também é mencionada no "Exit survey", indicando que a remuneração pode ser um fator relevante na satisfação dos funcionários, mas não é o único determinante.

Três maneiras diferentes de mostrar esses dados são:

## **Gráfico de Barras:** 

**Vantagens:**
Permite comparar facilmente as pontuações de "Ability to impact", "HRBP" e "Exit survey" para cada categoria, facilitando a identificação de diferenças exibindo a distribuição das pontuações em cada categoria.

**Deficiências:**
Pode ser difícil visualizar as diferenças sutis entre as barras quando há muitas categorias e valores baixos.

```python
# Definir a largura das barras
bar_width = 0.25

# Criar uma sequência de índices para posicionar as barras
index = np.arange(len(df))
plt.figure(figsize=(10, 9))
# Plotar as barras de Exit survey
plt.barh(index, df['Exit survey'], color='lightcoral', height=bar_width, label='Exit survey')

# Plotar as barras de HRBP
plt.barh(index + bar_width, df['HRBP'], color='lightgreen', height=bar_width, label='HRBP')

# Plotar as barras de Ability to impact
plt.barh(index + (2 * bar_width), df['Ability to impact'], color='skyblue', height=bar_width, label='Ability to impact')

# Personalizar o eixo y
plt.yticks(index + bar_width, df['Dados'])

# Personalizar o eixo x
plt.xlabel('Pontuação')
plt.title('Satisfação dos funcionários por categoria')

# Adicionar uma legenda
plt.legend()

# Adicionar os números das barras
for i, value in enumerate(df['Exit survey']):
    plt.text(value, i, str(value), ha='left', va='center', color='black')

for i, value in enumerate(df['HRBP']):
    plt.text(value, i + bar_width, str(value), ha='left', va='center', color='black')

for i, value in enumerate(df['Ability to impact']):
    plt.text(value, i + (2 * bar_width), str(value), ha='left', va='center', color='black')

# Mostrar o gráfico
plt.show()
```

### **Gráfico de Dispersão.** 

**Vantagens:**
Mostra a relação entre "Ability to impact" com "HRBP" e "Exit survey" para cada categoria.
Permite identificar padrões ou clusters nos dados.

**Deficiências:**
Pode ser difícil interpretar as pontuações exatas em cada ponto de dispersão.
Não fornece uma visão clara da distribuição das pontuações.


```python
# Gráfico de Dispersão 1
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Exit survey', y='Ability to impact', hue='Dados', s=100, palette='viridis')
plt.xlabel('Exit Survey')
plt.ylabel('Ability to impact')
plt.title('Relação entre Ability to impact e Exit Survey')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Gráfico de Dispersão 2

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='HRBP', y='Ability to impact', hue='Dados', s=100, palette='viridis')
plt.xlabel('HRBP')
plt.ylabel('Ability to impact')
plt.title('Relação entre Ability to impact e HRBP')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### **Gráfico de Linhas:**
**Vantagens:**
Permite acompanhar as tendências das pontuações ao longo das categorias.
Destaca facilmente as categorias com pontuações mais altas ou mais baixas.

**Deficiências:**
Não fornece uma representação visual precisa das pontuações individuais em cada categoria.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Dados', y='Ability to impact', marker='o', label='Ability to impact')
sns.lineplot(data=df, x='Dados', y='HRBP', marker='o', label='HRBP')
sns.lineplot(data=df, x='Dados', y='Exit survey', marker='o', label='Exit survey')
plt.xticks(rotation=35)
plt.xlabel('Dados')
plt.ylabel('Pontuação')
plt.title('Satisfação dos funcionários por categoria')
plt.legend()
plt.show()
```

A visualização que melhor representa os dados depende do objetivo da análise. Se o objetivo for comparar as métricas para cada categoria, o gráfico de barras agrupadas pode ser mais adequado. Se o objetivo for destacar a composição das pontuações em cada categoria, o gráfico de barras empilhadas pode ser preferível. Já se o objetivo for identificar possíveis relações entre as métricas de satisfação dos funcionários, o gráfico de dispersão pode ser mais adequado.
